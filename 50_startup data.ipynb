```python
plt.scatter(df['R&D Spend'], df['Profit'], color='blue', marker='o')

plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Spend vs Profit")
plt.show()
```

```python
x=df[["R&D Spend"]]
y=df["Profit"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
Lr = LinearRegression()         
Lr.fit(x_train, y_train)        
y_pred = Lr.predict(x_test)
```

```python
# FIND EVALUATION MATRIX IT HAS 2 
r2_score=r2_score(y_test,y_pred)
MSE=mean_squared_error(y_test,y_pred) #  its linear reg model ready
```

```python
import gradio as gr
import joblib
import numpy as np

# Load your trained model (Multiple Linear Regression / any model you trained)
model = joblib.load("C:\\Users\\hp\\Downloads.startpkl")

# Define prediction function
def predict_profit(RD_Spend, Administration, Marketing_Spend, State):
    # Handle categorical variable 'State'
    # Example: If you encoded 'State' during training, you must apply same encoding here
    # For simplicity, assume you used One-Hot Encoding with 3 states: 'New York', 'California', 'Florida'

    states = ['New York', 'California', 'Florida']
    state_encoded = [1 if State == s else 0 for s in states]

    # Combine features
    features = np.array([[RD_Spend, Administration, Marketing_Spend] + state_encoded])
    
    prediction = model.predict(features)
    return f"Predicted Profit: ${prediction[0]:,.2f}"

# Define Gradio interface
inputs = [
    gr.Number(label="R&D Spend"),
    gr.Number(label="Administration"),
    gr.Number(label="Marketing Spend"),
    gr.Dropdown(choices=["New York", "California", "Florida"], label="State"),
]

output = gr.Textbox(label="Predicted Profit")

gr.Interface(
    fn=predict_profit,
    inputs=inputs,
    outputs=output,
    title="Startup Profit Predictor"
).launch()
```
